```python
import asyncio
from telethon import TelegramClient
from telethon.tl.types import MessageMediaPhoto
import pandas as pd
import nest_asyncio
import os
from PIL import Image
import io
import logging

# Apply the nest_asyncio patch
nest_asyncio.apply()

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Telegram credentials
api_id = '27613793'
api_hash = '58f73f500af1aa0f525caff262964fa4'
phone_number = '+251911756372'

# Create a client with a custom session
session = 'my_session5'
client = TelegramClient(session, api_id, api_hash)

# Define the channels
channels = ['CheMed123', 'lobelia4cosmetics']

# Ensure output directory exists
output_dir = 'telegram_images_2'
os.makedirs(output_dir, exist_ok=True)

# Function to download images and log process
async def download_images(channels, image_limit):
    for channel_name in channels:
        channel = await client.get_entity(channel_name)
        async for message in client.iter_messages(channel, limit=image_limit):
            if message.media and isinstance(message.media, MessageMediaPhoto):
                try:
                    photo = await client.download_media(message.media, file=bytes)
                    image = Image.open(io.BytesIO(photo))
                    file_name = os.path.join(output_dir, f'{channel_name}_{message.id}.jpg')
                    image.save(file_name)
                    logging.info(f"Image saved to {file_name}")
                except Exception as e:
                    logging.error(f"Failed to download/save image: {e}")

# Main function to start client and run image download
async def main():
    logging.info("Current Working Directory: %s", os.getcwd())
    await client.start()
    await download_images(channels, image_limit=20)  # Adjust limit as needed
    await client.disconnect()

# Run the main function
asyncio.run(main())
```
